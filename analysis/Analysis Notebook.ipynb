# Research Question

- ***It is better to work in a small company than a larger one if you want to work in the field of Data Engineering?***

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
font_scale=1.5
```

```python
# Please note -1 is unknown Company Rating
fig, ax = plt.subplots(figsize=(15,10))
chart1 = sns.histplot(x="Company Rating",hue="Company Size",multiple="stack",data=df)
plt.title("Company Rating and Company Size Plot")
plt.ylabel("")
```

***From this graph, small company size seems to have a better company rating than medium and larger company***  

***Medium company still have good company rating***

***Company rating for large company is leaning toward average***

```python
# Please note -1 is unknown Company Rating
fig, ax = plt.subplots(figsize=(15,10))
chart2 = sns.histplot(x="Company Rating", 
                      y="Average Salary",
                      hue="Company Size",
                      hue_order=["Large","Medium","Small","Unknown"],
                      data=df)
plt.title("Company Rating and Average Salary Plot")
```

***From this graph, it seems that the majority of good company rating peaked at around the average salary 75000 and 110000 USD***

```python
#Company size and average salary
fig, ax = plt.subplots(figsize=(15,10))
chart3 = sns.histplot(x="Average Salary",hue="Company Size",multiple="stack",data=df)
```

***This dataset sample pool is reflected on which jobs were available on market at the time when this dataset was collected***

***That being said, in job market at that time, there were more jobs at a small company and the average salary is higher at a small company as well***

```python
fig, ax = plt.subplots(figsize=(20,10))
chart4 = sns.histplot(x="Company Size",hue="Location",multiple="stack",data=df)
```

***If you want to work for a small company then there are more jobs opening in New York***

***New York seems to be the place to go if you want a job in Data Engineering, Software Engineering, or Big Data Engineering in any company sizes***

# Answers to Research Question

- This data shows that small company has better company rating than larger company

- There are more people being content working at a small company with an average salary between 75000 to 110000 USD
